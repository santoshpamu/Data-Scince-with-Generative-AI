### Step-4:Split the data into train and test

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=0)
print('X_train.shape',X_train.shape)
print('X_test.shape',X_test.shape)
print('y_train.shape',y_train.shape)
print('y_test.shape',y_test.shape)
```

### Step-5:Apply  Data Preparation on Training data

```python
X_train.dtypes
```

##### Step 5.1: Separating numerical and categorical features

```python
X_train_num=X_train.select_dtypes(include=['int64','float64'])
X_train_cat=X_train.select_dtypes(include=['object'])                                 
```

```python
X_train_num.head()
```

```python
X_train_cat.head()
```

##### 5.2 Numerical feature transformation ("Standardization")

##### --> eventhough data is  in numerical format we perform standardization because to rescale every thing into same range


```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train_num_transformed=pd.DataFrame(sc.fit_transform(X_train_num),
                                     columns=sc.get_feature_names_out(),#column names are converted to numpyarray  if we use this columns names will be remain constant
                                     index=X_train_num.index)
X_train_num_transformed.head()
```

##### Lets analyse the Properties of each column using standardscaler

```python
print('number of numerical features:',    sc.n_features_in_)
print('ouput feature names:',             sc.get_feature_names_out())
print('mean of eaach column:',            sc.mean_)
print('std of each column:',              np.sqrt(sc.var_))
```

```python
X_train_num_transformed.round(2).head()
```

```python
cut_categories = ['Fair', 'Good', 'Very Good', 'Premium', 'Ideal']
color_categories = ['J', 'I', 'H', 'G', 'F', 'E', 'D']
clarity_categories = ['I1', 'SI2', 'SI1', 'VS2', 'VS1', 'VVS2', 'VVS1', 'IF']
```

##### 5.3 numerical feature transformation using Ordinal encoding

```python
from sklearn.preprocessing import OrdinalEncoder
ordinal=OrdinalEncoder(categories=[cut_categories,color_categories,clarity_categories],
                      handle_unknown='use_encoded_value',
                      unknown_value=-1,
                      encoded_missing_value=-5)
```

```python
X_train_cat_transformed=pd.DataFrame(ordinal.fit_transform(X_train_cat),
                                     columns=ordinal.get_feature_names_out(),
                                    index=X_train_cat.index)
X_train_cat_transformed.head()
```

```python

# Let's now analyse the properties of 'ordinal_encoder'

print("Applied encoding on:", ordinal.feature_names_in_)
print("Unique Categories:", ordinal.categories_)
print("Feature Names after encoding:", ordinal.get_feature_names_out())
```

### step-6: Concatenate X_train_num_transform and X_train_cat_transform

```python
X_train_transformed=pd.concat([X_train_num_transformed,X_train_cat_transformed],axis=1)
X_train_transformed.head()
```

#### Step-7: Applying Data Preparation on Test Data

```python
X_test_num=X_test.select_dtypes(include=['int64','float64'])
X_test_cat=X_test.select_dtypes(include=['object'])                                
```
